```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
#Generating points on a circle
def circ_gen(O,r):
	len = 50
	theta = np.linspace(0,2*np.pi,len)
	x_circ = np.zeros((2,len))
	x_circ[0,:] = r*np.cos(theta)
	x_circ[1,:] = r*np.sin(theta)
	x_circ = (x_circ.T + O).T
	return x_circ
```

```python
#Input parameters
#Intercepts along coordinate axis
a=2
b=2


C_1 = 1*np.array(([a/2,b/2])) #center (1,1)
f = 0
r_1 = np.sqrt(LA.norm(C_1)**2-f) #radius
x_circ1= circ_gen(C_1,r_1) #centre(1,1)
```

```python
x1 = [a] #Intercept along x-axis
y1 = [0] #Intercept along x-axis
x2 = [0] #Intercept along y-axis
y2 = [b] #Intercept along y-axis
c1=[1]  #center
c2=[1]  #center
o1=[0]  #origin
o2=[0]  #origin
plt.plot(x_circ1[0,:],x_circ1[1,:],label='$circle with centre(1,1)$')
x1,=plt.plot(x1, y1,marker='o',color='green',label='Point (2,0)')
x2,=plt.plot(x2, y2,marker='o',color='green',label='Point (0,2)')
c,=plt.plot(c1, c2,marker='o',color='black',label='Centre (1,1)')
o,=plt.plot(o1, o2,marker='o',color='orange',label='Origin (0,0)')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(bbox_to_anchor=(0.8, 1), loc='lower center')
plt.grid()
plt.axis('equal')
plt.savefig('circle1.png', bbox_inches='tight')
plt.show()
```
